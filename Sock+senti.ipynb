```python
import yfinance as yf
import requests
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import pandas_ta as ta
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.subplots as sp

# API keys for News API and Alpha Vantage
NEWS_API_KEY = ''
ALPHA_VANTAGE_API_KEY = ''

# Function to fetch news headlines and content related to a stock ticker from News API
def fetch_news_api_stock_news(ticker):
    url = f'https://newsapi.org/v2/everything?q={ticker}&apiKey={NEWS_API_KEY}'
    response = requests.get(url)
    news_data = response.json()
    articles = news_data.get('articles', [])
    headlines = [article['title'] for article in articles]
    contents = [article['content'] for article in articles]
    dates = [article['publishedAt'] for article in articles]
    return headlines, contents, dates

# Function to fetch news headlines and content related to a stock ticker from Alpha Vantage
def fetch_alpha_vantage_stock_news(ticker):
    url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={ticker}&apikey={ALPHA_VANTAGE_API_KEY}'
    response = requests.get(url)
    news_data = response.json()
    articles = news_data.get('feed', [])
    headlines = [article['title'] for article in articles]
    contents = [article['summary'] for article in articles]
    dates = [article['time_published'] for article in articles]
    return headlines, contents, dates

# Function to analyze sentiment of a given text
def analyze_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(text)
    return sentiment

# Function to categorize sentiment based on the compound score
def categorize_sentiment(compound_score):
    if compound_score >= 0.05:
        return "Positive"
    elif compound_score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Function to visualize sentiment analysis results by date
def plot_sentiment_analysis(sentiments, dates, ticker):
    pos = [sentiment['pos'] for sentiment in sentiments]
    neu = [sentiment['neu'] for sentiment in sentiments]
    neg = [sentiment['neg'] for sentiment in sentiments]
    compound = [sentiment['compound'] for sentiment in sentiments]

    # Convert dates to a suitable format for plotting
    x = np.arange(len(sentiments))

    plt.figure(figsize=(14, 7))
    plt.plot(x, pos, label='Positive', color='g')
    plt.plot(x, neu, label='Neutral', color='b')
    plt.plot(x, neg, label='Negative', color='r')
    plt.plot(x, compound, label='Compound', color='k', linestyle='--')
    plt.xlabel('Article Index')
    plt.ylabel('Sentiment Score')
    plt.title(f'Sentiment Analysis of {ticker} News Articles')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to visualize historical closing prices
def visualize_closing_prices(data, ticker):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data.index, y=data['Adj Close'], mode='lines', name='Adj Close'))
    fig.update_layout(title=f'{ticker} Adjusted Closing Prices', xaxis_title='Date', yaxis_title='Adjusted Close')
    fig.show()

# Function to fetch historical stock data
def fetch_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data

# Function to calculate technical indicators
def add_technical_indicators(data):
    data['SMA'] = ta.sma(data['Adj Close'], length=14)
    data['EMA'] = ta.ema(data['Adj Close'], length=14)
    data['RSI'] = ta.rsi(data['Adj Close'], length=14)
    data['MACD'] = ta.macd(data['Adj Close'])['MACD_12_26_9']
    bb = ta.bbands(data['Adj Close'], length=20)
    data['BB_high'] = bb['BBU_20_2.0']
    data['BB_mid'] = bb['BBM_20_2.0']
    data['BB_low'] = bb['BBL_20_2.0']
    data.dropna(inplace=True)
    return data

# Function to prepare data for Prophet
def prepare_prophet_data(data):
    df = data.reset_index()
    df = df.rename(columns={'Date': 'ds', 'Adj Close': 'y'})
    return df[['ds', 'y']]

# Function to forecast stock prices using Prophet
def forecast_with_prophet(data, periods=365):
    df = prepare_prophet_data(data)
    model = Prophet()
    model.fit(df)
    future = model.make_future_dataframe(periods=periods)
    forecast = model.predict(future)
    return forecast

# Function to visualize forecast with technical indicators
def visualize_forecast_with_indicators(data, forecast, ticker):
    fig = sp.make_subplots(rows=4, cols=1, shared_xaxes=True, vertical_spacing=0.02, subplot_titles=('Historical and Forecasted Prices', 'SMA and EMA', 'RSI', 'MACD'))

    # Plot Historical and Forecasted Prices
    fig.add_trace(go.Scatter(x=data.index, y=data['Adj Close'], mode='lines', name='Historical'), row=1, col=1)
    fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'], mode='lines', name='Forecast'), row=1, col=1)

    # Plot SMA and EMA
    fig.add_trace(go.Scatter(x=data.index, y=data['SMA'], mode='lines', name='SMA'), row=2, col=1)
    fig.add_trace(go.Scatter(x=data.index, y=data['EMA'], mode='lines', name='EMA'), row=2, col=1)

    # Plot RSI
    fig.add_trace(go.Scatter(x=data.index, y=data['RSI'], mode='lines', name='RSI'), row=3, col=1)

    # Plot MACD
    fig.add_trace(go.Scatter(x=data.index, y=data['MACD'], mode='lines', name='MACD'), row=4, col=1)

    fig.update_layout(
        title=f'{ticker} Stock Price Forecast with Technical Indicators',
        xaxis_title='Date',
        yaxis_title='Value'
    )
    fig.show()

# Function to calculate MSE, RMSE, and MAE
def calculate_metrics(data, forecast):
    actual_prices = data['Adj Close'].values
    forecasted_prices = forecast['yhat'].values[:len(actual_prices)]  # Match forecast to actual data length
    mse = mean_squared_error(actual_prices, forecasted_prices)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual_prices, forecasted_prices)
    return mse, rmse, mae

# Function to print summary statistics
def print_summary_statistics(sentiments):
    compound_scores = [sentiment['compound'] for sentiment in sentiments]
    print(f"Mean Compound Score: {np.mean(compound_scores)}")
    print(f"Median Compound Score: {np.median(compound_scores)}")
    print(f"Standard Deviation of Compound Score: {np.std(compound_scores)}")

# Main function to execute the workflow for multiple tickers
def main():
    tickers = ['AAPL', 'SHOP', 'MDLZ', 'KO']
    start_date = '2020-01-01'
    end_date = '2024-06-01'
    forecast_periods = 365  # Forecast for 1 year

    for ticker in tickers:
        print(f'\nProcessing {ticker}...\n')
        data = fetch_data(ticker, start_date, end_date)
        
        # Visualize historical closing prices
        visualize_closing_prices(data, ticker)
        
        # Add technical indicators
        data = add_technical_indicators(data)
        
        # Forecast with Prophet
        forecast = forecast_with_prophet(data, periods=forecast_periods)
        
        # Calculate metrics
        mse, rmse, mae = calculate_metrics(data, forecast)
        
        print(f'Metrics for {ticker}:')
        print(f'MSE: {mse}')
        print(f'RMSE: {rmse}')
        print(f'MAE: {mae}')
        
        # Visualize forecast with technical indicators
        visualize_forecast_with_indicators(data, forecast, ticker)

        # Fetch stock news headlines, content, and dates from both APIs
        news_api_headlines, news_api_contents, news_api_dates = fetch_news_api_stock_news(ticker)
        alpha_vantage_headlines, alpha_vantage_contents, alpha_vantage_dates = fetch_alpha_vantage_stock_news(ticker)

        # Combine headlines, contents, and dates from both sources
        headlines = news_api_headlines + alpha_vantage_headlines
        contents = news_api_contents + alpha_vantage_contents
        dates = news_api_dates + alpha_vantage_dates

        # Analyze sentiment for each headline and content, and store the results
        headline_sentiments = [analyze_sentiment(headline) for headline in headlines]
        content_sentiments = [analyze_sentiment(content) for content in contents if content]

        # Print and plot sentiment results for headlines
        print("Headline Sentiments:")
        for headline, sentiment, date in zip(headlines, headline_sentiments, dates):
            category = categorize_sentiment(sentiment['compound'])
            print(f"Date: {date}")
            print(f"Headline: {headline}")
            print(f"Sentiment: {sentiment} (Category: {category})")
            print("-" * 60)

        # Print summary statistics
        print_summary_statistics(headline_sentiments)

        # Plot sentiment analysis results
        plot_sentiment_analysis(headline_sentiments, dates, ticker)

        # Print and plot sentiment results for contents
        print("Content Sentiments:")
        for content, sentiment, date in zip(contents, content_sentiments, dates):
            if content:
                category = categorize_sentiment(sentiment['compound'])
                print(f"Date: {date}")
                print(f"Content: {content[:200]}...")  # Print first 200 characters for brevity
                print(f"Sentiment: {sentiment} (Category: {category})")
                print("-" * 60)

        # Print summary statistics for content sentiments if needed
        print_summary_statistics(content_sentiments)

if __name__ == "__main__":
    main()
```
